```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:/Users/Lenova/OneDrive/ML/StudentsPerformance.csv')
print(df)

# Set style
sns.set(style="whitegrid")

# 1. Score Distributions (Histogram)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
score_columns = ["math score", "reading score", "writing score"]

for i, col in enumerate(score_columns):
    sns.histplot(df[col], bins=20, kde=True, ax=axes[i], color="skyblue")
    axes[i].set_title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

```python
# 2. Average Scores by Gender (Bar Chart)
avg_scores_by_gender = df.groupby("gender")[score_columns].mean().reset_index()
avg_scores_melted = avg_scores_by_gender.melt(id_vars="gender", var_name="Subject", value_name="Average Score")

plt.figure(figsize=(8, 6))
sns.barplot(x="Subject", y="Average Score", hue="gender", data=avg_scores_melted, palette="coolwarm")
plt.title("Average Scores by Gender")
plt.show()
```

```python
# 5. Average Scores by Parental Level of Education (Bar Chart)
avg_scores_by_parent_edu = df.groupby("parental level of education")[score_columns].mean().reset_index()
avg_scores_melted_parent_edu = avg_scores_by_parent_edu.melt(
    id_vars="parental level of education", var_name="Subject", value_name="Average Score")

plt.figure(figsize=(12, 6))
sns.barplot(
    x="parental level of education",
    y="Average Score",
    hue="Subject",
    data=avg_scores_melted_parent_edu,
    palette="viridis",
)
plt.xticks(rotation=45)
plt.title("Average Scores by Parental Level of Education")
plt.show()
```

```python
# 4. Correlation Heatmap
correlation_matrix = df[score_columns].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Scores")
plt.show()
```

```python
# Count Plot: Parental Level of Education
plt.figure(figsize=(10, 6))
sns.countplot(x="parental level of education", data=df, palette="viridis", order=df["parental level of education"].value_counts().index)
plt.xticks(rotation=45)
plt.title("Count of Parental Level of Education")
plt.xlabel("Parental Education Level")
plt.ylabel("Count")
plt.show()
```

```python
# Boxplot for Scores
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[score_columns], palette="pastel")
plt.title("Boxplot of Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.show()
```

```python
# Pie Chart (Gender Distribution)
plt.figure(figsize=(6, 6))
df["gender"].value_counts().plot.pie(autopct="%1.1f%%", colors=["lightblue", "pink"], startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title("Gender Distribution")
plt.ylabel("")  # Hide y-label
plt.show()
```

```python
# Scatterplot (Math vs Reading Scores)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["math score"], y=df["reading score"], color="blue", alpha=0.6)
plt.title("Scatterplot of Math vs Reading Scores")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

```python
# Multiple Boxplots (Scores based on Gender)

sns.boxplot(x="gender", y="race/ethnicity", data=df, palette="coolwarm")
plt.title("race/ethnicity Distribution by Gender")
plt.show()
```

```python
#Boxplot: Scores Based on Parental Level of Education

sns.boxplot(x="parental level of education", y="math score", data=df, palette="coolwarm", order=df["parental level of education"].value_counts().index)
plt.xticks(rotation=45)
plt.title("Math Score Distribution by Parental Education Level")
plt.show()

sns.boxplot(x="parental level of education", y="reading score", data=df, palette="coolwarm", order=df["parental level of education"].value_counts().index)
plt.xticks(rotation=45)
plt.title("Reading Score Distribution by Parental Education Level")
plt.show()

sns.boxplot(x="parental level of education", y="writing score", data=df, palette="coolwarm", order=df["parental level of education"].value_counts().index)
plt.xticks(rotation=45)
plt.title("Writing Score Distribution by Parental Education Level")
plt.show()


```

```python
# 4. Boxplot: Scores Based on Test Preparation Course
sns.boxplot(x="test preparation course", y="math score", data=df, palette="Set2")
plt.title("Math Scores by Test Preparation Course")
plt.show()

sns.boxplot(x="test preparation course", y="reading score", data=df, palette="Set2")
plt.title("Reading Scores by Test Preparation Course")
plt.show()

sns.boxplot(x="test preparation course", y="writing score", data=df, palette="Set2")
plt.title("Writing Scores by Test Preparation Course")
plt.show()
```
